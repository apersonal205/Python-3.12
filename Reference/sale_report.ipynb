# Data Analysis

```python
df.columns
```

```python
gender_counts = df['Gender'].value_counts() #Get
print("No of Orders by Gender")

plt.figure(figsize=(10, 10))
offset_fraction = 0.1 
custom_labels = ['Male', 'Female', 'Other'] 
explode = tuple(offset_fraction if i == 0 else 0 for i in range(len(gender_counts)))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90,explode=explode)
plt.title('No of Orders by Gender')
```

```python
ax : Axes = sns.countplot(x = 'Age Group',data = df, hue = 'Gender')
print("No of Orders by Age with Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Age Group', ascending=True)
# print(sales_age)

print(f"Total Amount by Age Group")

ax = sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group')

# Add labels to the bars
for idx, value in enumerate(sales_age['Amount']):
    ax.text(idx, value, value, ha='center', va='bottom')
    



    
```

```python
order_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='State', ascending=True).head(10)
# print(sales_age)

print(f"No of Order by State")
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(x = 'State',y= 'Orders' ,data = order_state, hue = 'Orders')

# Add labels to the bars
for idx, value in enumerate(order_state['Orders']):
    ax.text(idx, value, value, ha='center', va='bottom')

```

```python
sale_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True).head(10)

print(f"No of Amount by State")
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(x = 'State',y= 'Amount' ,data = sale_state, hue = 'Amount')

# Add labels to the bars
for idx, value in enumerate(sale_state['Amount']):
    ax.text(idx, value, value, ha='center', va='bottom') 

```

```python
occupation_counts = df['Occupation'].value_counts()
print("No of Orders by Occupation")

plt.figure(figsize=(10, 10))
offset_fraction = 0.1 
explode = tuple(offset_fraction if i == 0 else 0 for i in range(len(occupation_counts)))
plt.pie(occupation_counts, labels=occupation_counts.index, autopct='%1.1f%%', startangle=90,explode=explode)
plt.title('Distribution of Orders by Occupation')
```
